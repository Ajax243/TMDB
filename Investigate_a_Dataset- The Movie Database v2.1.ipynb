## Data Wrangling



```python
# Loading the data
df=pd.read_csv('tmdb-movies.csv')

df.head()
```

Choosing columns to drop that has a lot of missing values and unstructured data

```python
columns_to_drop=['id','imdb_id','homepage','overview','keywords','tagline']
```

```python
df.drop(columns=columns_to_drop, inplace=True)
```

 Checking the null and missing values in the remaining columns

```python
df.info()
```

<b>Many aspects of the analysis will be focused on budget and revenue, if they are missing or equal to zero they will create problems when trying to calculate the means, so will have to drop all the values that don't provide insight regarding those features. </b>

```python
df[(df['cast'].isnull()) & (df['revenue']== 0)]
```

Most of the null values in the cast are of foreign or movies that are not relevant with a lot of missing values

```python
df.dropna(subset= ['cast'] ,inplace=True)
```

```python
df.info()
```

```python
df.dropna(subset= ['director'] ,inplace=True)
```

Checking and dropping duplicated values

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

Production companies have around 1000 missing value, but they will not be relevant to every scope.                         
So I will change the null values to 'unknown' in case the scope leads to them .                                                                            
<b>But first I have to check that no production company has 'unknown' in their names.</b>

```python
df['production_companies'].str.contains('unknown').sum()
```

Same goes for the genres

```python
df['production_companies'].fillna('unknown', inplace=True)
df['genres'].fillna('unkown',inplace=True)
```

```python
df.info()
```

<b>Will introduce a new column that will be an indicator for the financial success of the movies called 'ratio' </b>

```python
df['ratio']=df['revenue']/df['budget']
```

Reordering the columns

```python
df=df[['original_title','release_year','vote_average','popularity','ratio', 'budget', 'revenue', 'cast', 'director',
       'runtime', 'genres', 'production_companies', 'release_date',
       'vote_count', 'budget_adj',
       'revenue_adj']]
```

<strong>Converting the budget and revenue columns to millions instead of dollars to be easier to read </strong>

```python
df['budget']=df['budget']/1000000
df['revenue']=df['revenue']/1000000

df['budget_adj']=df['budget_adj']/1000000
df['revenue_adj']=df['revenue_adj']/1000000
```

<br><b> From now on, all the values will be in Million USD</b> <br>

```python
df['ratio'].describe()
```

```python
df['ratio'].value_counts()
```

# After inspecting the 'ratio' column along with budget and revenue: <br>
- Around 2200 values either have missing values in budget or revenue. <br>
- Many movies either have clearly wrong values or being a 'No budget movie' which is by definition a movie with a budget of less than 400k USD<br>
- Since I am using the Revenue not the adjusted revenue I will include movies with budgets starting 250k USD and revenue of over 200k USD<br>

- The decision to use actual budgets and actual revenues is better in my opinion because it will discuss the numbers as facts in line of the scope of the analysis I will do

- Adjusted revenue and adjusted budget will be used only when calculating all-time values


```python
df1=df[(df['budget']>=.25) & (df['revenue']>= 0.2)]
```

```python
df1['release_date']=pd.to_datetime(df1['release_date'],format='%x')
```


### Data Cleaning


- There will be a problem with the 'cast' column. <br>
- We can't take just the first name either
- In some cases like 'Titanic', Leonardo Decaprio's name came second so he won't be counted <br>
<br>
# For this another dataframe needs to be made where the movies are duplicated each time with the actor's name is the name in the cast

At a later step there was a problem faced due to missing values in the names of the second cast members. <br>

Since all the names are in one column, I will create a copy, divide the names into 'cast' , 'second' and 'third
while keeping the rest of the names in the column 'third'  <br>

I am doing this to get the indices of the records with missing second and third names


```python
df2=df1.copy()
```

```python
df2[['cast','second','third']]=df2['cast'].str.split('|',2, expand=True)
```

```python
ind_2=df2[df2['second'].isnull()==True].index
```

```python
ind_2
```

We can drop these two values

```python
df1.drop(ind_2,inplace=True)
```

```python
df2.head()
```

# Now that there are not any missing values in the second and third names

```python
df21=df1.copy()
df22=df1.copy()
df23=df1.copy()
```

Splitting the cast column into 3 separate dataframes, each with the name of one of the first three cast members in the 'cast' column

```python
split_columns=['cast']
for c in split_columns:
    df21[c]=df21[c].apply(lambda x: x.split("|")[0])
    df22[c]=df22[c].apply(lambda x: x.split("|")[1])
    df23[c]=df23[c].apply(lambda x: x.split("|")[2])
```

```python
df23.shape
```

Putting them back together in one Dataframe

```python
df3=df21.append(df22)
df3=df3.append(df23)
```

### Research Question 2 : Do movies released in specific months perform better than average? If yes, which months?

The mean and median for the revenue per a movie in this dataset is:

```python
rev_mean=df1['revenue'].mean()
rev_median=df1['revenue'].median()

print('The mean revenue per movie is: ',float(rev_mean))
print('The median revenue per movie is: ', rev_median)
```

Which indicates that the distribution is highly skewed to the right <br>
<br>
<br>
Will take a look at the distribution

```python
plt.hist(df1['revenue'], bins=50);
plt.xlabel('Revenue in Million USD')
plt.ylabel('Count of movies')
plt.title('The distribution of revenues per movies')
```

<b>Lets look at the average revenue per month</b>

```python
month=range(1,13,1)
def target(par1):
    y=[]
    for i in range(1,13,1):
           y.append(df1[df1['release_date'].dt.month==i][par1].mean())    
    
    return y
plt.bar(month,target('revenue'))
plt.xlabel('Month')
plt.ylabel('Revenue in Million USD')
plt.title('Average Revenue per Month')
plt.show()
```

<b>There seems to be a clear pattern that June comes first, then May then November then December <br>
    But this does not mean that Movies released these months have higher revenue on average.<br>
    <br>
    It could mean : <br>
    - That movies released at these times have higher budgets (which could be related to revenue, which we will get into later) <br>
    - There are more movies in specific months that affects the mean count and create a fake pattern
</b>

This is the average budget per month

```python
plt.bar(month,target('budget'))
plt.xlabel('Month')
plt.ylabel('Budget in Million USD')
plt.title('Average Budget per Month')
plt.show()
```

<br> Almost the same pattern is observed. But this could mean that movies in these dates have a higher or a lower count that might be skewing the average <br>

<b>Lets look at the count of movies released each month</b>

```python
color = sns.color_palette()[0]
df1['month'] = df1['release_date'].dt.month
sns.countplot(x=df1['month'], color = color );
plt.xlabel('Months');
plt.ylabel('Count of Movies released');
plt.title('Number of Releases by Month');
```

<b> Completely different pattern, actually it shows that on November there should be the higher revenue, which means that movies in september are less profitable than average</b> <br>
<br>
<b> Lets take a look at the profitability, which is the ratio between the revenue and the budget of the movie</b>

```python
plt.plot(month,target('ratio'), marker ='o')
plt.xlabel('Month')
plt.ylabel('Revenue/Budget')
plt.title('Mean Profitability of Movies for each Month')
plt.show()
```

<b> This partially makes sense that November has the lowest profitability, with the highest count, creating the least revenue still. <br>
    But on October shows and average count, highest profitability with a low revenue <br>
    <br>
    This could be because of outliers, lets remove outliers based on 1.5 IQR and plot it again: </b>

```python
df_plot = df1[(df1['ratio']>= -2.3) & (df1['ratio']<= 7.035)]
ratio_per_month=[]
for i in range(1,13,1):    
    ratio_per_month.append(df_plot[df_plot['release_date'].dt.month==i]['ratio'].mean())

plt.plot(month,ratio_per_month, marker ='o')
plt.xlabel('Month')
plt.ylabel('Revenue/Budget')
plt.title('Average Profitability of Movies for each Month')
plt.show()
```

<b> Now this makes perfect sense, which describes why movies in June and December have higher profitability <br>
    </b> <br>
    Lets plot the above graph with median instead of mean to compare with median values:

```python
def target2(par1):
    x=[]
    for i in range(1,13,1):    
        x.append(df1[df1['release_date'].dt.month==i][par1].median())
    return x

plt.plot(month,target2('ratio'), marker ='o')
plt.xlabel('Month')
plt.ylabel('Ratio (Revenue/Budget)')
plt.title('Median Profitability of Movies for each Month')
plt.show()
```

BONUS CHECK:<br>
Almost identical which begs the question, is the median a better representation than the mean for these values <br>
Lets look at the average revenue per month looking at the median instead of the mean:

```python
plt.bar(month,target2('revenue'))
plt.xlabel('Month')
plt.ylabel('Median Revenue')
plt.title('Median Revenue for each Month')
plt.show()
```

### Research Question 3 : Do movies with higher budgets tend to be rated higher or be more popular?

<b>Lets look at pearson correlation coefficient along with the plot between budget and average vote :</b>

```python
def relation(x1,y1):
    plt.scatter(x=df1[x1], y=df1[y1]);
    plt.xlabel(x1.title())
    plt.ylabel(y1.title())
    plt.title('The Relationship between {} and {}'.format(x1.title(),y1.title()))
    print('The Correlation is: ')
    return df1[x1].corr(df1[y1], method='pearson')
relation('budget','vote_average')
```

<b> There seems to  no correlation between budget and the average vote on IMDB </b>

<b>Lets look at pearson correlation coefficient along with the plot between budget and popularity :</b>

```python
relation('budget','popularity')
```

<b>There seems to be a moderate correlation between the Budget of the movie and its Popularity</b>
